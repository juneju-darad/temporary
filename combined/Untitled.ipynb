```python
import glob
read_files = glob.glob("*.txt")
with open("merged.txt", "wb") as outfile:
    for f in read_files:
        with open(f, "rb") as infile:
            outfile.write(infile.read())
```

```python
with open("merged.txt", "r") as s:
    lines = s.readlines()
with open("merged.txt", "w") as s:
    for line in lines:
        if line.strip("\n") != "Node Name	Level	Type":
            s.write(line)
```

```python
import pandas as pd
txt = pd.read_table('merged.txt', names=('Node_Name', 'Level', 'Type'),dtype={'Node_Name': str, 'Level': str,'Type': str})
```
